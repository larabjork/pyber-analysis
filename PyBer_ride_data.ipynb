```python
# Import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
# Load in csv
ride_data  = os.path.join("Resources", "PyBer_ride_data.csv")
pyber_ride_df = pd.read_csv(ride_data)
pyber_ride_df
```

```python
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations
# This is an array that is the length of the DF
x_axis = np.arange(len(pyber_ride_df))
# This uses list comprehension to set the x-axis tick locs
tick_locations = [value for value in x_axis]
# Plot the data
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)")
# This passes tick_locations which has the values of the array, as labels for x-ticks
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# Make it a bar chart, by chaining .bar method
pyber_ride_df.plot.bar(x = "Month", y = "Avg. Fare ($USD)")
plt.show()
```

```python
# Make it a bar chart, by adding kind param to plot function
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)", kind = "bar")
plt.show()
```

```python
# Skill Drill
import statistics


# Change bar colors to sky blue
pyber_ride_df.plot.bar(x = "Month", 
                       y = "Avg. Fare ($USD)",
                       color = "xkcd:sky blue", )
# Add error bars
# Add caps to error bars
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])
plt.errorbar(pyber_ride_df["Month"], 
              pyber_ride_df["Avg. Fare ($USD)"],
              yerr  = stdev,
              capsize = 3)
# Rotate labels on x-axis to horizontal
plt.tick_params(axis='x', rotation=0)
# Set the y-axis increment to $5
plt.yticks(np.arange(0, 51, step = 5.0))
plt.show()
```
