```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total far in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with the MATLAB method
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt(), which is the object-oriented interface method
fig, ax = plt.subplots() #this creates a figure (fig references the figure)
ax.plot(x_axis, y_axis) #this defines the axes(ax referneces the figure and allows annotations)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
# with only the two lines above, you get an empty chart
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Using plt.show(), which looks for all active figure objects and opens a window that displays them if there are 2+ sets of data to plot
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#5.1.4 - Annotating MATLAB charts
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = "Boston")
# Create labels for x, y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label = "Boston")
# Create labels for x, y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
#Annotating O-O Interface Charts

# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
# with only the two lines above, you get an empty chart
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label = "Boston")
# Create labels for x, y axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
# Set the y limit between 0 and 45
ax.set_ylim(0, 45)
# Create a title
ax.set_title("PyBer Fare by Month")
# Add a grid
ax.grid()
# Add the legend
ax.legend()
```

```python
# 5.1.5. Create bar charts with MATLAB approach
# Vertical bar chart

# Create the plot with .bar function
plt.bar(x_axis, y_axis)
```

```python
# Create the plot with .bar function
plt.bar(x_axis, y_axis, color = "green", label = "Boston")
# Create labels for x, y axes
plt.xlabel("Date")
plt.ylabel("Fares $")
# Create a title
plt.title("PyBer Fare by Month")
# Add a legend
plt.legend()
```

```python
# Horizontal bar chart with .barh function
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Swap axes (i.e., so dates read left right not up/down)
plt.barh(y_axis, x_axis)
```

```python
# Invert order of dates, so January at top, Dec at bottom
# Create the plot
plt.barh(x_axis, y_axis)
# use gca method to "get current axis" then "invert_yaxis"
plt.gca().invert_yaxis()
```

```python
# Annotate the axes
# Create the plot
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
# use gca method to "get current axis" then "invert_yaxis"
plt.gca().invert_yaxis()
# add title and axes labels
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fares ($)")
# add a legend
plt.legend()
```

```python
# 5.1.6 Bar charts with O-O Approach
# Vertical bar chart
# Create plot with ax.subplots()
fix, ax = plt.subplots()
# Use .bar() to create bar chart
ax.bar(x_axis, y_axis)
```

```python
# Horizontal bar chart
# Create the plot with .subplots()
fig, ax = plt.subplots()
# Create horizontal bar chart with barh()
ax.barh(x_axis, y_axis)
```

```python
# Swap x, y positions, even though it's not helpful for this data
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# Skill drill
# Change bar color to cyan
# Switch so that Fares is on x axis
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")
# Add legend for city of Chicago
ax.legend()
# Add title and axes labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fares ($)")
ax.set_ylabel("Date")
# Invert y-axis data
ax.invert_yaxis()
```
